5 Nama Mahasiswa

```python
mahasiswa = ["yanto", "rusdi", "vergil", "mas eko", "kakang"]
print("list nama mahasiswa :")
print(mahasiswa)
```

Tambahkan 1 nama baru dengan append

```python
mahasiswa.append("Dekde")
print("Setelah ditambahkan nama baru")
print(mahasiswa)
```

Hapus satu element menggunakan remove

```python
mahasiswa.remove("vergil")
print("\nSetelah menghapus nama vergil:")
print(mahasiswa)
```
